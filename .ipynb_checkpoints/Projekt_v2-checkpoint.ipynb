First, we are downloading the csv data and doing explorative data analysis from AT emissions to check which categories are fitting for the car data. Since the data is from the years 1990-2022, and the car data is from 2000, just the emissions from 2000 are being considered. First, we get the CO2 data from the emissions csv. Here we are looking at the line 33, which has the yearly CO2 data from transport. Now we write us a script which gets just the line we want (in our case line 33, for transport)

```python
import pandas as pd

# Load the CSV file with a specified delimiter
data = pd.read_csv('emissions/schadstoffemissionen_1990-2022_nach_nfr_wide_1a3b1.csv', delimiter=';')

# Apply the filter for 'NFR_Code' equals '1 A 3 b 1' and 'Schadstoff' equals 'CO'
filtered_data = data[(data['NFR_Code'] == '1 A 3 b 1') & (data['Schadstoff'] == 'CO')]

# Print the number of rows to see how many entries we have
print("Number of entries matching '1 A 3 b 1' and 'CO':", filtered_data.shape[0])

# Proceed if there are any rows after filtering
if filtered_data.shape[0] > 0:
    # Access the first row, or any specific row, as needed
    year_data = filtered_data.iloc[0]  # Adjust .iloc[0] to another index if needed

    # Create a DataFrame with just this row of year data
    cleaned_data = pd.DataFrame([year_data])

    # Write the DataFrame to a new Excel file
    cleaned_data.to_excel('emissions/cleaned_co_emissions_1a3b1.xlsx', index=False)

    print("Excel file has been created with the required data.")
else:
    print("No data matches the specified filters.")
```

we need a little bit more cleanup

```python
import pandas as pd

def clean_excel_data(input_path, output_path):
    # Load the original Excel file
    data = pd.read_excel(input_path)
    
    # Keep only columns with year data, starting from 1990 to the latest available year
    year_columns = [col for col in data.columns if col.isdigit() and int(col) >= 1990]
    cleaned_data = data[year_columns]
    
    # Convert data types to float, ensuring that commas are treated as decimal separators
    cleaned_data = cleaned_data.apply(lambda x: x.str.replace(',', '.').astype(float))
    
    # Save the cleaned data to a new Excel file
    cleaned_data.to_excel(output_path, index=False)


# Paths to the files (update these paths as necessary)
input_file_path = 'emissions/cleaned_co_emissions_1a3b1.xlsx'
output_file_path = 'emissions/cleaned_co_emissions.xlsx'

# Call the function to clean the data
clean_excel_data(input_file_path, output_file_path)
```

# Used car data and correlation

```python

import pandas as pd
import matplotlib.pyplot as plt

# Load the data
file_path = 'usedcars/overalldata_used.csv'
data = pd.read_csv(file_path)

# Split the single column into separate columns by the semicolon delimiter
data_split = data['C-J59-0;C-A10-0;C-EK7-0;F-ISIS-1'].str.split(';', expand=True)

# Rename the columns based on the assumed headers from the original column names
data_split.columns = ['C-J59-0', 'C-A10-0', 'C-EK7-0', 'F-ISIS-1']

# Convert 'F-ISIS-1' to numeric for summation
data_split['F-ISIS-1'] = pd.to_numeric(data_split['F-ISIS-1'], errors='coerce')

# Extract year from 'C-A10-0'
data_split['Year'] = data_split['C-A10-0'].str.slice(4, 8).astype(int)

# Filter the DataFrame to include only the years from 2000 to 2022
filtered_data = data_split[(data_split['Year'] >= 2000) & (data_split['Year'] <= 2022)]

# Group by 'Year' and sum up 'F-ISIS-1'
yearly_sum = filtered_data.groupby('Year')['F-ISIS-1'].sum().reset_index()

# Display the result
print(yearly_sum)


# Assuming 'yearly_sum' has already been created with the above code block you provided
# Load the emissions data
cleaned_data_path = 'emissions/cleaned_co_emissions.xlsx'  # Replace with the correct path to the cleaned data
cleaned_data = pd.read_excel(cleaned_data_path)

# Filter the data for the years 2000 to 2022
years = list(range(2000, 2023))  # These are the column names in the emissions data
austria_co2 = cleaned_data[years].transpose()

# Rename the index and columns for clarity in the emissions data
austria_co2.index.name = 'Year'
austria_co2.columns = ['CO2 Emissions']

# Ensure 'Year' in both datasets is the same dtype, ideally int for proper alignment
yearly_sum['Year'] = yearly_sum['Year'].astype(int)

# Merge the two datasets on the 'Year' column
combined_data = austria_co2.join(yearly_sum.set_index('Year'), how='inner')

# Plot the data
plt.figure(figsize=(14, 7))
plt.subplot(1, 2, 1)
plt.plot(combined_data.index, combined_data['CO2 Emissions'], marker='o', label='CO2 Emissions (tons)')
plt.title('CO2 Emissions in Austria (2000-2022)')
plt.xlabel('Year')
plt.ylabel('CO2 Emissions (tons)')
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(combined_data.index, combined_data['F-ISIS-1'], marker='o', color='red', label='Used Car Data Metric')
plt.title('Used Car Data Metric (2000-2022)')
plt.xlabel('Year')
plt.ylabel('Metric Value')
plt.grid(True)
plt.legend()

plt.tight_layout()

# Save the plot as a PNG file
graph_path = 'emissions/austria_co2_emissions_2000_2022.png'  # Replace with your desired file path
plt.savefig(graph_path)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load and prepare the used car data
file_path = 'usedcars/overalldata_used.csv'
data = pd.read_csv(file_path)
data_split = data['C-J59-0;C-A10-0;C-EK7-0;F-ISIS-1'].str.split(';', expand=True)
data_split.columns = ['C-J59-0', 'C-A10-0', 'C-EK7-0', 'F-ISIS-1']
data_split['F-ISIS-1'] = pd.to_numeric(data_split['F-ISIS-1'], errors='coerce')
data_split['Year'] = data_split['C-A10-0'].str.slice(4, 8).astype(int)
filtered_data = data_split[(data_split['Year'] >= 2000) & (data_split['Year'] <= 2022)]
yearly_sum = filtered_data.groupby('Year')['F-ISIS-1'].sum().reset_index()

# Load and prepare the CO2 emissions data
cleaned_data_path = 'emissions/cleaned_co_emissions.xlsx'
cleaned_data = pd.read_excel(cleaned_data_path)
years = list(range(2000, 2023))
austria_co2 = cleaned_data[years].transpose()
austria_co2.index = austria_co2.index.astype(int)  # Convert index to integer for alignment
austria_co2.index.name = 'Year'
austria_co2.columns = ['CO2 Emissions']

# Merge the two datasets on the 'Year' column
combined_data = yearly_sum.merge(austria_co2, on='Year', how='inner')

# Calculate and display the Pearson correlation coefficient
correlation = combined_data[['F-ISIS-1', 'CO2 Emissions']].corr()
print("Correlation Matrix:\n", correlation)

# Plot the combined data to visualize relationships
plt.figure(figsize=(10, 5))
plt.scatter(combined_data['CO2 Emissions'], combined_data['F-ISIS-1'])
plt.title('Scatter Plot of Used Car Data Metric vs. CO2 Emissions')
plt.xlabel('CO2 Emissions (tons)')
plt.ylabel('Used Car Data Metric (Sum of F-ISIS-1)')
plt.grid(True)
plt.show()

# Save the combined data table for reference
combined_data.to_csv('combined_data_correlation.csv', index=False)
```

    A coefficient close to 1 indicates a strong positive correlation, meaning as one variable increases, the other also increases.
    A coefficient close to -1 indicates a strong negative correlation, meaning as one variable increases, the other decreases.
    A coefficient close to 0 suggests no linear correlation between the variables.

## Fragestellung: Ãndert sich die Korrelation, wenn man beide Werte von gebraucht und Neuwagen aufsummiert?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load and prepare the new car data
new_car_path = 'newcars/overalldata.csv'
new_data = pd.read_csv(new_car_path)
new_data_split = new_data['C-J59-0;C-A10-0;C-EK7-0;F-ISIS-1'].str.split(';', expand=True)
new_data_split.columns = ['C-J59-0', 'C-A10-0', 'C-EK7-0', 'F-ISIS-1']
new_data_split['F-ISIS-1'] = pd.to_numeric(new_data_split['F-ISIS-1'], errors='coerce')
new_data_split['Year'] = new_data_split['C-A10-0'].str.slice(4, 8).astype(int)
new_filtered_data = new_data_split[(new_data_split['Year'] >= 2000) & (new_data_split['Year'] <= 2022)]
new_yearly_sum = new_filtered_data.groupby('Year')['F-ISIS-1'].sum().reset_index()

# Load and prepare the used car data
used_car_path = 'usedcars/overalldata_used.csv'
used_data = pd.read_csv(used_car_path)
used_data_split = used_data['C-J59-0;C-A10-0;C-EK7-0;F-ISIS-1'].str.split(';', expand=True)
used_data_split.columns = ['C-J59-0', 'C-A10-0', 'C-EK7-0', 'F-ISIS-1']
used_data_split['F-ISIS-1'] = pd.to_numeric(used_data_split['F-ISIS-1'], errors='coerce')
used_data_split['Year'] = used_data_split['C-A10-0'].str.slice(4, 8).astype(int)
used_filtered_data = used_data_split[(used_data_split['Year'] >= 2000) & (used_data_split['Year'] <= 2022)]
used_yearly_sum = used_filtered_data.groupby('Year')['F-ISIS-1'].sum().reset_index()

# Combine new and used car data
combined_yearly_sum = pd.merge(new_yearly_sum, used_yearly_sum, on='Year', suffixes=('_new', '_used'))
combined_yearly_sum['Total F-ISIS-1'] = combined_yearly_sum['F-ISIS-1_new'] + combined_yearly_sum['F-ISIS-1_used']

# Load and prepare the CO2 emissions data
cleaned_data_path = 'emissions/cleaned_co_emissions.xlsx'
cleaned_data = pd.read_excel(cleaned_data_path)
years = list(range(2000, 2023))
austria_co2 = cleaned_data[years].transpose()
austria_co2.index = austria_co2.index.astype(int)  # Convert index to integer for alignment
austria_co2.index.name = 'Year'
austria_co2.columns = ['CO2 Emissions']

# Merge the combined car data with CO2 emissions data
final_combined_data = combined_yearly_sum.merge(austria_co2, on='Year', how='inner')

# Calculate and display the Pearson correlation coefficient
correlation = final_combined_data[['Total F-ISIS-1', 'CO2 Emissions']].corr()
print("Correlation Matrix:\n", correlation)

# Plot the combined data to visualize relationships
plt.figure(figsize=(10, 5))
plt.scatter(final_combined_data['CO2 Emissions'], final_combined_data['Total F-ISIS-1'])
plt.title('Scatter Plot of Total Car Data Metric vs. CO2 Emissions')
plt.xlabel('CO2 Emissions (tons)')
plt.ylabel('Total Car Data Metric (Sum of F-ISIS-1)')
plt.grid(True)
plt.show()

# Save the combined data table for reference
final_combined_data.to_csv('final_combined_data_correlation.csv', index=False)
```

since the value is negative, there is a somewhat negative correlation --> the car values and emission data do not correlate

### Get new emission data which reflects passenger cars

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
data = pd.read_excel('emissions/cleaned_co_emissions_1a3b1.xlsx')

# Focus on the years 2000-2022, assuming column names are like '2000', '2001', ..., '2022'
years = [str(year) for year in range(2000, 2023)]
kt_values = data[years].values.flatten()  # Extract data for the specified years

# Create a plot
plt.figure(figsize=(14, 7))
plt.subplot(1, 2, 1)
plt.plot(years, kt_values, marker='o')
plt.title('CO Emissions from 2000 to 2022 (1 A 3 b 1)')
plt.xlabel('Year')
plt.ylabel('Emissions (kt)')
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
